```python
# prompt: Generate a Python script for a Jupyter notebook that simulates customer purchase behavior in a retail industry scenario. The simulation should involve 10,000 customers and 100 different products.

import random
import numpy as np
import pandas as pd

# Define number of customers and products
num_customers = 10000
num_products = 100

# Create a list of product IDs
product_ids = [i for i in range(1, num_products + 1)]

# Create a list of customer IDs
customer_ids = [i for i in range(1, num_customers + 1)]

# Generate random purchase data
purchase_data = []
for customer_id in customer_ids:
  num_purchases = random.randint(1, 5)  # Each customer buys 1 to 5 products
  for _ in range(num_purchases):
    product_id = random.choice(product_ids)
    purchase_amount = round(random.uniform(10, 500), 2)  # Purchase amount between 10 and 500
    purchase_data.append([customer_id, product_id, purchase_amount])

# Create a Pandas DataFrame from the purchase data
df_purchases = pd.DataFrame(purchase_data, columns=['Customer ID', 'Product ID', 'Purchase Amount'])

# Analyze the data (optional)
# Example: Calculate total revenue per product
total_revenue_per_product = df_purchases.groupby('Product ID')['Purchase Amount'].sum()
print(total_revenue_per_product)

# Example: Calculate average purchase amount per customer
average_purchase_per_customer = df_purchases.groupby('Customer ID')['Purchase Amount'].mean()
print(average_purchase_per_customer)

# You can perform further analysis based on your requirements
# For instance, you can analyze customer demographics, product categories,
# purchase frequency, seasonality, etc.
```

```python
# prompt: Include the following: Generate random customer IDs and product IDs.
# Create random purchase quantities for each customer-product pair using a Poisson distribution.
# Store the data in a Pandas Dataframe
# Calculate the total purchases per customer and per product.
# Visualize the distribution of total purchases per customer and per product using histograms.

import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define number of customers and products
num_customers = 10000
num_products = 100

# Create a list of product IDs
product_ids = [i for i in range(1, num_products + 1)]

# Create a list of customer IDs
customer_ids = [i for i in range(1, num_customers + 1)]

# Generate random purchase data with purchase quantities
purchase_data = []
for customer_id in customer_ids:
  for product_id in product_ids:
    purchase_quantity = np.random.poisson(lam=2)  # Poisson distribution with lambda=2
    if purchase_quantity > 0:  # Only add if there is a purchase
      purchase_amount = round(random.uniform(10, 500), 2) * purchase_quantity
      purchase_data.append([customer_id, product_id, purchase_quantity, purchase_amount])

# Create a Pandas DataFrame from the purchase data
df_purchases = pd.DataFrame(purchase_data, columns=['Customer ID', 'Product ID', 'Purchase Quantity', 'Purchase Amount'])


# Calculate total purchases per customer
total_purchases_per_customer = df_purchases.groupby('Customer ID')['Purchase Amount'].sum()

# Calculate total purchases per product
total_purchases_per_product = df_purchases.groupby('Product ID')['Purchase Amount'].sum()


# Visualize the distribution of total purchases per customer using a histogram
plt.figure(figsize=(10, 5))
plt.hist(total_purchases_per_customer, bins=50)
plt.xlabel('Total Purchases per Customer')
plt.ylabel('Number of Customers')
plt.title('Distribution of Total Purchases per Customer')
plt.show()


# Visualize the distribution of total purchases per product using a histogram
plt.figure(figsize=(10, 5))
plt.hist(total_purchases_per_product, bins=50)
plt.xlabel('Total Purchases per Product')
plt.ylabel('Number of Products')
plt.title('Distribution of Total Purchases per Product')
plt.show()
```
